```python
# .. 메뉴추가 모듈 만들기
# (내용) 4가지 파일 리스트에 새로운 메뉴(단어) 추가
# 


# .. 수정파일 리스트
# 1. user_dic.tsv: 형태소 분리용, 사용자 추가 단어
# 2. corpus.txt: 학습용(intent 모델), 사전용(ner 모델)
#                 ner_train.txt 문장을 포함하고 있음
# 3. total_train_data.csv: corpus.txt랑 똑같음, 단지 csv
# 4. ner_train.txt: copurs.txt에서 문장들을 BIO로 구분해놓은 자료
import re
from re import search, findall, match, sub


def readFile(file):

    sentences = None
    
    try:
        print(f'open {file}')
        with open(file, mode= 'r', encoding = 'UTF8') as f:
            sentences = f.readlines()
            #print(f'{file} opened successfully')
    except:
        pass

    finally:
        return sentences


def find_refference(ref="짜장면", sents="None"):

    num = 0
    ref_lines = []
    for i in sents :

        if search(ref, i) and i:   # search 는 문자열의 중간에 '짜장면'이 위치할 때도 성공함 
            num +=1
            ref_lines.append(i)
            #print(i)
        else:
            pass

    print(f'참조 메뉴: {ref}, 문장개수: ',num)

    return ref_lines


def replace_ref2newmenu(ref_word="짜장면", ref_lines=None, new_word=None):

    new_lines = []
    for i in ref_lines:
        new_lines.append(i.replace(ref_word, new_word))

    print(f"{len(new_lines)}개 치환 완료")
    
    return new_lines


def create_new_file(new_data, new_file):

    try:
        with open(new_file, mode= 'w', encoding = 'UTF8') as f:
            for line in new_data:
                f.write(f"{line}")
            print("새파일 작성 완료")
    except Exception as e:
            print(e)



file_list = ["./user_dic.tsv",
             "../train_tools/dict/corpus.txt",
             "../model/intent/total_train_data.csv",
             "../model/ner/ner_train.txt"]

file_nlist = ["./user_dic1.tsv",
              "../train_tools/dict/corpus1.txt",
              "../model/intent/total_train_data1.csv",
              "../model/ner/ner_train1.txt"]


for a, b in zip(file_list, file_nlist):

    print("="*80)
    #print(a,b)
    sentences = readFile(a)
    #print("sentences: ", sentences[0:5])
    #print(type(sentences))
    ref_lines = find_refference(sents=sentences)
    new_lines = replace_ref2newmenu(ref_lines=ref_lines, new_word="싱싱회")
    #print("new_lines", new_lines[0:5])
    sentences.extend("\n")
    sentences.extend(new_lines)
    #print(sentences[-5:-1])
    create_new_file(new_data=sentences, new_file=b)


```
