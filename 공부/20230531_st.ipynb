* class = 특정 기능을 하는 프레임
  * attributes & method
* instance = class의 실제적용 유닛, class로 찍어낸것
* 객체 = 자신의 고유한 속성을 가지고 있어, 다른것들과 식별 가능한것
* 특정 클래스의 인스턴스가 생겨 고유한 속성을 가지면 객체라고 말할 수 있는거 같음
* 보통 인스턴스와 객체를 혼용하여 사용
* overiding = 상속
* overloading = 중복 함수


* 동영상강의
  * 동적인웹 서버
    * 사용자에게 정보를 요청할 수 있는 페이지를 발송하고
    * 사용자가 정보를 요청하면, 요청정보를 반영한 웹페이지를 다시 리턴
      * 사용자의 신호를 적을수 있는 칸을 html에 따로 만들어서 그거에 대한 응답을 웹서버에 전송
    * 요즘에는 보통 클라우드로 서버를 구성
    * 웹서버를 구축시 필요한 서버들
      * 정적 서버(앞단 배치)
        * 아파치
      * 동적 서버(WAS; 웹애플리케이션서버, 뒷단 배치)
        * 톰캡, 장고
      * DB 서버(뒷단 배치)
          * 구매기록 등 정보 저장
          * oracle: DB 1위 기업
