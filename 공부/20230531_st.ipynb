* class = 특정 기능을 하는 프레임
  * attributes & method
* instance = class의 실제적용 유닛, class로 찍어낸것
* 객체 = 자신의 고유한 속성을 가지고 있어, 다른것들과 식별 가능한것
* 특정 클래스의 인스턴스가 생겨 고유한 속성을 가지면 객체라고 말할 수 있는거 같음
* 보통 인스턴스와 객체를 혼용하여 사용
* overiding = 상속
* overloading = 중복 함수

* 동영상강의2
  * 데이터베이스 테이블
    * 행,열(속성)
    * 유니크한 컬럼을 주로 키로 만듦(ex, 회원번호)
  * DBMS
    * 사용자 인터페이스, 데이터베이스언어(SQL)로 구성
    * MySQL
    * SQLite3
  * SQLite3
    * 기기 내부에서만(외부접근 X)
    * 파이썬에서는 SQLite3 기본 내장
    * 혼자쓰기에 완벽한 DB는 아님
  * 테이블 만들기
  ```sqlite3
      Create table [if not exists] table-name("name" "type" "constraint")
  ```
  * CRUD 사용법
    * Create - 생성 - insert
      * ```insert into table (column-name) values;```
    * Read - 읽기/조회 - select
      * ```Select column(또는 *) from table where 조건;```
    * Update - 수정/갱신 - update
      * ```update table set column = 수정값 where 조건식;```
    * Delete - 삭제 - delete
      * ```Delete from table-name where 조건식;```
